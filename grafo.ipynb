```python
import networkx as nx
import matplotlib.pyplot as plt
import random

# Criação do grafo direcionado com 12 vértices
G = nx.DiGraph()

# Adicionando os vértices
vertices = [f'V{i}' for i in range(1, 13)]
G.add_nodes_from(vertices)

# Adicionando arestas com pesos aleatórios entre os vértices
arestas = [
    ('V1', 'V2'), ('V2', 'V3'), ('V3', 'V4'), ('V4', 'V5'),
    ('V5', 'V6'), ('V6', 'V7'), ('V7', 'V8'), ('V8', 'V9'),
    ('V9', 'V10'), ('V10', 'V11'), ('V11', 'V12'), ('V12', 'V1'),
    ('V1', 'V5'), ('V2', 'V6'), ('V3', 'V7'), ('V4', 'V8'),
    ('V5', 'V9'), ('V6', 'V10'), ('V7', 'V11'), ('V8', 'V12'),
    ('V9', 'V1'), ('V10', 'V2'), ('V11', 'V3'), ('V12', 'V4')
]

# Adicionando arestas com pesos aleatórios entre 1 e 10
for u, v in arestas:
    G.add_edge(u, v, weight=random.randint(1, 10))

# Escolhendo um caminho para destacar (exemplo: de V1 a V6)
caminho_destacado = ['V1', 'V5', 'V6']

# Gerando a visualização
plt.figure(figsize=(10, 8))

# Definindo o layout
pos = nx.spring_layout(G, seed=42)

# Desenhando o grafo
edges = G.edges()
weights = nx.get_edge_attributes(G, 'weight')

# Desenhando as arestas com pesos
nx.draw_networkx_edges(G, pos, edgelist=edges, width=1.0, alpha=0.7, edge_color='gray')

# Desenhando as arestas do caminho destacado em vermelho
highlighted_edges = [(caminho_destacado[i], caminho_destacado[i+1]) for i in range(len(caminho_destacado)-1)]
nx.draw_networkx_edges(G, pos, edgelist=highlighted_edges, width=3, edge_color='red')

# Desenhando os nós
nx.draw_networkx_nodes(G, pos, node_size=2000, node_color='lightblue')

# Desenhando os rótulos dos nós
nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

# Desenhando os pesos das arestas
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10)

# Título
plt.title("Grafo Direcionado com Pesos e Caminho Destacado (V1 -> V6)", fontsize=14)

# Exibindo o grafo
plt.axis('off')
plt.show()
```
